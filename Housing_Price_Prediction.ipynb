```python
hp.drop(['PoolQC','MiscFeature','Alley','Fence','FireplaceQu','LotFrontage'], axis = 1, inplace = True)
```

```python
#Checking for total null values
hp.isnull().sum()[hp.isnull().sum()>0].sort_values(ascending=False)
```

```python
# Removed ID column as it was not important for training algorithm.
hp.drop(['Id'],axis=1, inplace=True)

# MSSubClass is actually a categorical variable, hence converting it to the categorical one
hp['MSSubClass'] = hp['MSSubClass'].astype('object')
```

```python
# Finding How many Numeric and Categorical features
numeric_features = hp.select_dtypes(exclude=['object'])#.copy()
categorical_features = hp.select_dtypes(include=['object'])#.copy()

print("Number of numerical features are: ", numeric_features.shape[1])
print("Number of categorical features are: ", categorical_features.shape[1])
```

```python
# Printing Numeric and Categorical features
numeric_features.columns, categorical_features.columns
```

```python
#checking for outliers on Numeric features
plt.figure(figsize = (25,100))
count = 1
for col in numeric_features:
    plt.subplot(20,4,count)
    plt.boxplot(numeric_features[col])
    plt.title(col)
    count +=1
        
plt.show()
```

```python
# converting categorical data into Numerical data
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for col in categorical_features.columns:
    hp[col]=labelencoder.fit_transform(hp[col])
hp.head() 
```

```python
hp.tail()
```

```python
#seperating X and y
X = hp.iloc[:,:-1].values
y = hp.iloc[:,-1].values
```

```python
#dividing into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
MLR = LinearRegression()
MLR.fit(X_train,y_train)
```

```python
#prediction
y_pred = MLR.predict(X_test)
print(MLR.coef_)
print(MLR.intercept_)
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
```

```python
prediction_output =pd.DataFrame(data=[y_test,y_pred],index=['y_test','y_pred'])
prediction_output.transpose()
```
